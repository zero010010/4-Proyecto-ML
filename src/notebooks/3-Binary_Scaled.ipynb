```python
import pandas as pd 
import numpy as np  
import matplotlib.pyplot as plt
import os
```

```python
df=pd.read_csv('/Users/miguelopez/Desktop/4-Proyecto ML/src/data_raw/Carbon Emission.csv')
```

```python
# Missing Values, vehicle type  

df['Vehicle Type'].fillna('unknown', inplace=True)
```

```python
df.isna().sum()
```

```python
df.isnull().any()
```

```python
# Numerical and categorical features 

numerical = []
categorical = []

for col in df.columns:
    if df[col].dtype == 'int64':
        numerical.append(col)
    elif df[col].dtype == 'object':
        categorical.append(col)

print("Categorical: ",categorical)
print("Numerical: ",numerical)
```

```python
df.tail()
```

```python
df.columns
```

```python
# Create two dataframes containing numerical and categorical columns
df_numerical = df[numerical]
df_categorical = df[categorical]    

df_numerical.head()
```

```python
df_numerical.tail()
```

```python
# MinMaxScaler()

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df_numerical_scaled = df_numerical.copy()  
df_numerical_scaled[df_numerical.columns] = scaler.fit_transform(df_numerical)
```

```python

df_numerical_scaled
```

```python
df_categorical
```

```python
# Binary encoding of categorical variables
from ast import literal_eval

data = {
  'Body Type': ['overweight', 'obese', 'underweight', 'normal'],
  'Sex': ['female', 'male', 'female', 'male'],
  'Diet': ['pescatarian', 'vegetarian', 'omnivore', 'vegan'],
  'How Often Shower': ['daily', 'less frequently', 'more frequently', 'twice a day'],
  'Heating Energy Source': ['coal', 'natural gas', 'wood', 'electricity'],
  'Transport': ['public', 'walk/bicycle', 'private', 'public'],
  'Vehicle Type': ['unknown', 'petrol', 'diesel', 'hybrid'],
  'Social Activity': ['often', 'never', 'sometimes', 'often'],
  'Frequency of Traveling by Air': ['frequently', 'rarely', 'never', 'very frequently'],
  'Waste Bag Size': ['large', 'extra large', 'small', 'medium'],
  'Energy efficiency': ['No', 'Sometimes', 'Yes', 'No'],
  'Recycling': ["['Metal']", "['Paper', 'Plastic', 'Glass', 'Metal']", "['Paper']", "['Paper', 'Glass', 'Metal']"],
  'Cooking_With': ["['Stove', 'Oven']", "['Stove', 'Microwave']", "['Oven', 'Microwave']", "['Microwave', 'Grill', 'Airfryer']"]
}

df_dummies = pd.DataFrame(data)

# Recycling and Cooking_With columns to lists
df_dummies['Recycling'] = df_dummies['Recycling'].apply(literal_eval) 
df_dummies['Cooking_With'] = df_dummies['Cooking_With'].apply(literal_eval)

# Flatten lists to strings
df_dummies['Recycling'] = df_dummies['Recycling'].apply(lambda x: ','.join(x))
df_dummies['Cooking_With'] = df_dummies['Cooking_With'].apply(lambda x: ','.join(x))

# dummy columns 
df_dummies = pd.get_dummies(df_dummies)


```

```python
df_dummies.head(20)
```

```python
# Merge scaled numerical data with dummies  
df = df_numerical_scaled.merge(df_dummies, left_index=True, right_index=True)
# Save final df to csv as "df_numerical_scaled_dummies.csv"
df.to_csv('df_numerical_scaled_dummies.csv')
```
